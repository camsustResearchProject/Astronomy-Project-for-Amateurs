```python
from astropy.io import fits
from astropy.visualization import simple_norm
import matplotlib.pyplot as plt
import numpy as np

# File path for the FITS file
fits_file = 'RGB IMage1.fits'
fits_data = fits.open(fits_file)
fits_data.info()

# Extracting individual channels from the FITS file
red_data = fits_data[1].data
green_data = fits_data[2].data
blue_data = fits_data[3].data

# Applying normalization to each color channel
norm_red_func = simple_norm(red_data, 'sqrt', percent=99.5)
norm_green_func = simple_norm(green_data, 'sqrt', percent=99.5)
norm_blue_func = simple_norm(blue_data, 'sqrt', percent=99.5)

red_norm_data = norm_red_func(red_data)
green_norm_data = norm_green_func(green_data)
blue_norm_data = norm_blue_func(blue_data)

# Creating empty arrays for each color filter image
red_filtered_img = np.zeros((red_data.shape[0], red_data.shape[1], 3), dtype=float)
green_filtered_img = np.zeros((green_data.shape[0], green_data.shape[1], 3), dtype=float)
blue_filtered_img = np.zeros((blue_data.shape[0], blue_data.shape[1], 3), dtype=float)

# Assigning normalized data to the appropriate channels
red_filtered_img[:, :, 0] = red_norm_data  # Red data to red channel
green_filtered_img[:, :, 1] = green_norm_data  # Green data to green channel
blue_filtered_img[:, :, 2] = blue_norm_data  # Blue data to blue channel

# Writing the filtered images back to FITS files
red_hdu = fits.PrimaryHDU(red_filtered_img)
red_hdulist = fits.HDUList([red_hdu])
red_hdulist.writeto('Red_Filtered.fits', overwrite=True)

green_hdu = fits.PrimaryHDU(green_filtered_img)
green_hdulist = fits.HDUList([green_hdu])
green_hdulist.writeto('Green_Filtered.fits', overwrite=True)

blue_hdu = fits.PrimaryHDU(blue_filtered_img)
blue_hdulist = fits.HDUList([blue_hdu])
blue_hdulist.writeto('Blue_Filtered.fits', overwrite=True)

# Checking and printing minimum and maximum values for each channel
print("Red channel min/max:", red_data.min(), red_data.max())
print("Green channel min/max:", green_data.min(), green_data.max())
print("Blue channel min/max:", blue_data.min(), blue_data.max())

# Creating an RGB composite image using the normalized data
rgb_composite_img = np.zeros((red_data.shape[0], red_data.shape[1], 3), dtype=float)
rgb_composite_img[:, :, 0] = red_norm_data  # Assigning normalized red data
rgb_composite_img[:, :, 1] = green_norm_data  # Assigning normalized green data
rgb_composite_img[:, :, 2] = blue_norm_data  # Assigning normalized blue data

# Displaying the RGB composite image
plt.figure()
plt.imshow(rgb_composite_img)
plt.title('Composite RGB Image')
plt.show()

# Displaying individual filter images

# Red filtered image
plt.figure()
plt.imshow(red_filtered_img)
plt.title('Red Filtered Image')
plt.show()

# Green filtered image
plt.figure()
plt.imshow(green_filtered_img)
plt.title('Green Filtered Image')
plt.show()

# Blue filtered image
plt.figure()
plt.imshow(blue_filtered_img)
plt.title('Blue Filtered Image')
plt.show()
```
