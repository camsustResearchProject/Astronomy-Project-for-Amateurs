#### `Matplotlib`

```python
import matplotlib.pyplot as plt

# data
time = np.array([0, 1, 2, 3, 4, 5])  
distance = np.array([10, 12, 14, 11, 9, 5])

# plotting
plt.scatter(time, distance, color = 'red', marker='+')
plt.show()
```

```python
# labeling

time = np.array([0, 1, 2, 3, 4, 5])  
distance_1 = np.array([10, 12, 14, 11, 9, 5])  
distance_2 = np.array([10, 12.5, 15, 10, 8, 3.5])

plt.plot(time, distance_1, label = 'Without Air Resistance')
plt.plot(time, distance_2, label = 'With Air Resistance')
plt.title('Pendulam Distance')
plt.xlabel('Time in second')
plt.ylabel('Distance in meter')
plt.legend()
plt.show()
```

```python
# trig

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()
```

```python
#subplot

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)

plt.figure(figsize=(10,6)) # creating the canvas
plt.subplot(211)           # 2 row 1 col position - 1
plt.plot(x, y_1)           
plt.title('$\\sin{x}$')
plt.subplot(212)           # 2 row 1 col position - 2 
plt.plot(x, y_2)
plt.title('$\\cos{x}$')
plt.show()
```
